```python
import random

class Player():
    def __init__(self, symbol):
        self.symbol = symbol
        print(f"Enter the name of the Player ({self.symbol}):")
        self.name = input()
        print(f"Welcome to the game {self.name}.")
    
    def __repr__(self):
        return f"{self.name} ({self.symbol})"

```

```python
# Asks player to chose a cell and checks if it's a valid move
def player_turn(player, board):
    while True: #not sure why?
        cell = int(input(f"{player}, select a cell (1-9): ")) - 1  # Index start with 0, so we adjust it with -1
        if board[cell] == ' ' and 0 <= cell < 9:
            return cell
        else:
            print("You can't do this. Please choose an empty cell (1-9).")
```

```python
# Checks if a player won the game
def check_win(player, board):
    winning_combinations = [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Rows
        (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Columns
        (0, 4, 8), (2, 4, 6)              # Diagonals
    ]
    for combo in winning_combinations:
        if all(board[i] == player.symbol for i in combo):
            return True
    return False
```
