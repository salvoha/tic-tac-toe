```python
#summary
import random
class Player:
    def __init__(self, name, symbol):
        self.name = name
        self.symbol = symbol
#creating players
def create_players():
    name1 = input("Player A, please enter your Name: ").strip() or "Player A"
    name2 = input("Player B, please enter your Name: ").strip() or "Player B"

    while True:
        symbol1 = input(f"{name1}, choose your symbol (X/O): ").strip().upper()
        if symbol1 in ("X", "O"):
            break
        print("Invalid choice. Please type X or O.")

    if symbol1 == "X":
        symbol2 = "O" 
    else:
        symbol2 = "X"

    player1 = Player(name1, symbol1)
    player2 = Player(name2, symbol2)

    return player1, player2
#making a move
def make_a_move(board,player):
    while True:
        move = input(f'{player} {player}, choose a Position (1-9): ')
        #is input no number??
        if not move.isdigit():
            print("Please enter a number between 1 and 9.")
            continue
        #cut 1 to get the index and check if in range
        pos = int(move) - 1
        if pos < 0 or pos > 8:
            print("Position must be between 1 and 9.")
            continue
        #check if field is taken or not
        if board[pos] != " ":
            print("That field is already taken.")
            continue

        return pos

    

#is board full?
def is_board_full(board):
    return " " not in board


```

```python
#how to win?
WINNING_COMBINATIONS = [
    (0, 1, 2), #top row
    (3, 4, 5), #middle row
    (6, 7, 8), #down row
    (0, 3, 6), #left cloumn
    (1, 4, 7), #middl column
    (2, 5, 8), #right column
    (0, 4, 8), #diagonal left to right
    (2, 4, 6), #diagonal right to left
]

def check_win(board, symbol):
    for a, b, c in WINNING_COMBINATIONS:
        if board[a] == board[b] == board[c] == symbol:
            return True
        else:
            return False
```
